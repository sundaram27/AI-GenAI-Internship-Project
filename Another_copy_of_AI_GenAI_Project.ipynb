```python
from transformers import pipeline
```

```python
text_generator = pipeline("text-generation", model="gpt2")
```

```python
prompt = "Artificial Intelligence is transforming industries because"
output = text_generator(prompt, max_length=60, num_return_sequences=1)

print(output[0]['generated_text'])
```

# Evaluation

The model is able to generate coherent text from a short prompt.  
It produces meaningful sentences, although sometimes the results may drift off-topic.  
This shows both the usefulness and current limitations of generative AI models.
